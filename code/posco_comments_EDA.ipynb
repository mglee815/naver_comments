```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("../result/POSCO_comment_sentiment.csv")
```

```python
df.groupby('sentiment').size()
```

```python
df.groupby('media').size().sort_values(ascending = False)[:10]
```

```python
df['c_day'] = df.c_time.apply(lambda x : x[9:11])
```

```python
parameters = {'xtick.labelsize': 30,
          'axes.titlesize': 30,
          'figure.figsize' : (60, 30),
          'figure.dpi' : 200
          }
plt.rcParams.update(parameters)
plt.plot(df.groupby('c_day').size().sort_index())
plt.title('# of Comments in SEP')
```

```python
plt.plot(df.groupby('c_day').sum()['sentiment'])
```
